```
from google.colab import drive
import os
import errno
from pathlib import Path
import numpy as np
import cv2

face_cascade = cv2.CascadeClassifier('/usr/local/lib/python3.6/dist-packages/cv2/data/haarcascade_frontalface_alt2.xml')
drive.mount( '/content/drive' )

def ls( ruta = Path.cwd() ):
    return [ arch.name for arch in Path(ruta).iterdir() if arch.is_file() ]
```

```
with open( "/content/drive/My Drive/Colab Notebooks/TPFinal/dataset.txt", "r" ) as file:
  line = file.readline()
  while ( line != '' ) :
    ruta  = "/content/drive/My Drive/Colab Notebooks/TPFinal/deteccionCaras/dataSet/" + str( line[:4] ) + '/' + str( line[5:8] ) +'/' 
    os.makedirs(ruta)
    line = file.readline()
    #lista = ls(ruta)
```

```
with open( "/content/drive/My Drive/Colab Notebooks/TPFinal/dataset.txt", "r" ) as file:
  line = file.readline()
  while ( line != '' ) :
    ruta = "/content/drive/My Drive/Cohn-kanade-dataset/" + str( line[:4] ) + '/' + str( line[5:8] ) + '/'
    lista = ls(ruta)
    print( ruta )
    for arch in lista:
      img = cv2.imread( ruta + arch )
      gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
      faces = face_cascade.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=5, minSize=(30, 30),flags=cv2.CASCADE_SCALE_IMAGE)
      for (x,y,w,h) in faces:
        facedata = img[y:y+h, x:x+w]
        cv2.imwrite("/content/drive/My Drive/Colab Notebooks/TPFinal/deteccionCaras/dataSet/"+ str( line[:4] ) + '/' + str( line[5:8] ) + '/' + arch , facedata)
    line = file.readline()
```
