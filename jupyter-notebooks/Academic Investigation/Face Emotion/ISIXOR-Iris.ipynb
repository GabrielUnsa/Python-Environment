Aplicado al DataSet XOR

```

import numpy as np
import math

X = np.array( ( [ 0, 0 ], [ 0, 1 ], [ 1, 0 ], [ 1, 1 ] ), dtype=float )
y = np.array( ( [ 0 ], [ 1 ], [ 1 ], [ 0 ] ), dtype=float )
```

Modelo para XOR

```

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD


model = Sequential()
model.add( Dense( 10, input_dim = 2, activation = 'relu' ) )
model.add( Dense( 1, activation = 'sigmoid' ) )

epocas = 100
lrate = 0.01
decay = lrate/epocas
sgd = SGD( lr = lrate, momentum = 0.9, decay = decay, nesterov = False )

model.compile( loss = 'binary_crossentropy', optimizer = sgd, metrics=['accuracy'] )

model.fit( X, y, epochs=epocas, batch_size = 1 )
scores = model.evaluate( X, y )
print("\n%s: %.2f%%" %( model.metrics_names[1], scores[1] * 100 ) )
```

Aplicado al dataset Iris

```

from google.colab import drive
import numpy as np

drive.mount('/content/gdrive')

dataset = np.loadtxt("/content/gdrive/My Drive/DataSets/data.txt",delimiter=',')
```

```

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD

y_train = []
y_test = []
for i in range( 150 ):  
  if( i <= 50):
    if ( i >= 40 and i < 50 ):
      y_test.append( [ 1, 0, 0 ] )
    else:
      y_train.append( [ 1, 0, 0 ] )
  elif( i <= 100 ):
    if( i >= 90 and i < 100 ):
      y_test.append( [ 0, 1, 0 ] )
    else:
      y_train.append( [ 0, 1, 0 ] )
  else:
    if( i >= 140 and i < 150 ):
      y_test.append( [ 0, 0, 1 ] )
    else:
      y_train.append( [ 0, 0, 1 ] )

X_train = np.concatenate( ( dataset[:40,:4], dataset[50:90,:4], dataset[100:140,:4] ) )
X_train /= np.amax( X_train, axis = 0 )
y_train = np.array( y_train, dtype=float )

X_test = np.concatenate( ( dataset[40:50,:4], dataset[90:100,:4], dataset[140:150,:4] ) )
X_test /= np.amax( X_test, axis = 0 )
y_test = np.array( y_test, dtype=float )

model = Sequential()
model.add( Dense( 8, input_dim = 4, activation = 'relu' ) )
model.add( Dense( 3, activation = 'sigmoid' ) )

epocas = 50
lrate = 0.01
decay = lrate/epocas
sgd = SGD( lr = lrate, momentum = 0.9, decay = decay, nesterov = False )

model.compile( loss = 'categorical_crossentropy', optimizer = sgd, metrics=['accuracy'] )

model.fit(X_train, y_train, epochs=epocas, batch_size = 1 )
scores = model.evaluate( X_test, y_test )
print("\n%s: %.2f%%" %( model.metrics_names[1], scores[1] * 100 ) )
```
