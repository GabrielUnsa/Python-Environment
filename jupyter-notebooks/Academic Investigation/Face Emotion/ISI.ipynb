```
from google.colab import drive
import numpy as np

drive.mount('/content/gdrive')
#uploaded = files.upload()

dataset = np.loadtxt("/content/gdrive/My Drive/DataSets/pima-indians-diabetes.csv",delimiter=',')
X_train = dataset[:600,0:8]
y_train = dataset[:600,8]

X_test = dataset[600:,0:8]
y_test = dataset[600:,8]
```

```
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD

X_train = X_train / np.amax( X_train, axis = 0 )
X_test = X_test / np.amax( X_test, axis = 0 )

model = Sequential()
model.add( Dense( 100, input_dim = 8, activation = 'relu' ) )
model.add( Dense( 75, activation = 'relu' ) )
model.add( Dense( 1, activation = 'sigmoid' ) )

epocas = 200
lrate = 0.01
decay = lrate/epocas
sgd = SGD( lr = lrate, momentum = 0.9, decay = decay, nesterov = False )

model.compile( loss = 'binary_crossentropy', optimizer = sgd, metrics=['accuracy'] )

model.fit(X_train, y_train, epochs=epocas, batch_size = 1 )
scores = model.evaluate( X_test, y_test )
print("\n%s: %.2f%%" %( model.metrics_names[1], scores[1] * 100 ) )
```
