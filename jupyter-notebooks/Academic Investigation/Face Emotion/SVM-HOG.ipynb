```

import cv2
import numpy as np
from google.colab import drive
drive.mount('/content/gdrive')


hog = cv2.HOGDescriptor()


X_train, y_train = list(), list()
X_test, y_test   = list(), list()


for l in range(1,11):
    for i in range(1,9):
        if ( l != 10):
            ruta = '/content/gdrive/My Drive/DataSets/huellas/10' + str(l) + '_' + str(i) + '.tif'
        else:
            ruta = '/content/gdrive/My Drive/DataSets/huellas/1' + str(l) + '_' + str(i) + '.tif'
        img = cv2.imread(ruta, cv2.IMREAD_COLOR)
        descriptor = hog.compute(img).ravel()
        if(i < 7):
            X_train.append(descriptor)
            y_train.append(l)
        else:
            X_test.append(descriptor)
            y_test.append(l)

X_train = np.array(X_train, dtype = np.float32)
y_train = np.array(y_train, dtype = np.int32)

X_test = np.array(X_test, dtype = np.float32)
y_test = np.array(y_test, dtype = np.int32)

SVM = cv2.ml.SVM_create()
SVM.setType(cv2.ml.SVM_C_SVC)
SVM.setDegree(3)
SVM.setKernel(cv2.ml.SVM_POLY) #Polinomial
SVM.setTermCriteria((cv2.TERM_CRITERIA_COUNT, 100, 1.e-6))
SVM.train(X_train, cv2.ml.ROW_SAMPLE, y_train)

res, y_predict = SVM.predict(X_test[2:4])
if( y_test[2] == (int) (y_predict[0][0]) and y_test[3] == (int)(y_predict[1][0])):
  print("El sujeto es: %d"%y_predict[0][0])
else:
  print("Concentrese en los sujetos: {}{}".format(y_predict[0][0],y_predict[1][0]))
  
res, y_predict = SVM.predict(X_test[:2])
if( y_test[0] == (int) (y_predict[0][0]) and y_test[1] == (int)(y_predict[1][0])):
  print("El sujeto es: %d"%y_predict[0][0])
else:
  print("Concentrese en los sujetos: {}, {}".format((int)(y_predict[0][0]),(int)(y_predict[1][0])))
  
res, y_predict = SVM.predict(X_train[25:27])
if( y_train[25] == (int) (y_predict[0][0]) and y_train[26] == (int)(y_predict[1][0])):
  print("El sujeto es: %d"%y_predict[0][0])
else:
  print("Concentrese en los sujetos: {}, {}".format((int)(y_predict[0][0]),(int)(y_predict[1][0])))
```
