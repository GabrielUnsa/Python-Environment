```
#Este modulo nos permite conectar con una webcam de manera remota para la deteccion del rostro
'''
Para poder funcionar necesitamos instalar en el celu(NO NECESARIAMENTE SOLO FUNCIONA CON EL CELU)
la apk IP WEBCAM y empezar a transmitir luego conn la ip que nos da la webcam podemos 
hacer un constate request para poder obtener la imagen lo codificamos (cosa que hay que ver si es necesario)
y abrimos la imagen
'''
import requests
import urllib as ul
import cv2
import numpy as np 

url = 'http://192.168.100.3:8080/shot.jpg'

while True:
    imgResp = ul.request.urlopen(url)
    imgNp = np.array(bytearray(imgResp.read()), dtype=np.uint8)
    img = cv2.imdecode(imgNp,1)
    cv2.imshow('test',img)
    cv2.waitKey(10)
```

```
## http://sefiks.com/2018/01/10/real-time-facial-expression-recognition-on-streaming-data/
import numpy as np
import cv2
# haarcascade_frontalface_alt.xml se encuentra con archivos 
face_cascade = cv2.CascadeClassifier('/usr/local/lib/python3.6/dist-packages/cv2/data/haarcascade_frontalface_alt.xml')

cap = cv2.VideoCapture(0)
 
while(True):
    # ret valor booleano si la camara esta disponible
    #img pixel capturados
    ret, img = cap.read()
    
    #apply same face detection procedures
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
 
    for (x,y,w,h) in faces:
        #print(faces)
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2) # cuadrado azul
        cv2.imshow( 'img', img )
    if cv2.waitKey(1) & 0xFF == ord('q'): #presionar q to quit
        break
        
cap.release()
cv2.destroyAllWindows() 
```
