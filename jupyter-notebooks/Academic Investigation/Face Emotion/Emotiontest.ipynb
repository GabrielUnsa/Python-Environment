```
import matplotlib
import matplotlib.pyplot as plt

def show_image_and_label(x, y):
    x_reshaped = x.reshape(48,48)
    plt.imshow(x_reshaped, cmap= "gray",
              interpolation="nearest")
    plt.axis("off")
    plt.show()
    print(y)
```

```
import numpy as np
# x_pixels
img = raw_data["pixels"][150]
val = img.split(" ")
x_pixels = np.array(val, 'float32')
x_pixels /= 255
```
