```python
import numpy as np

class graph_ma:
  def __init__( self ):
    self.matrix = None
    self.poblation = None
    self.solution = None
    self.nVertex = None
    self.nPoblation = None
    self.generation = None

  def viewMatrix( self ):
    for edge in self.matrix:
      if 1 in edge:
        print( edge )

  # Load edge datas
  def readData( self, path ):
    with open( path, 'r' ) as f:
      line = f.readline()
      self.nVertex = int( line.split(' ')[2] ) + int( line.split(' ')[3] )
      self.matrix = np.zeros( ( (self.nVertex - 1) , (self.nVertex - 1) ) )
      line = f.readline()
      while line != '':
        edge = line.split( ' ' )
        self.matrix[ int(edge[1]) - 1 ][ int(edge[2]) - 1 ] = 1
        self.matrix[ int(edge[2]) - 1 ][ int(edge[1]) - 1 ] = 1
        line = f.readline()
    f.close()

  #Solucion inicial
  def initSolution( self, valCol ):
    self.nPoblation = random.randint( 2, random.randint( 2, 10 ) )
    self.solution = np.random.randint( valCol, size = self.nVertex ) 
    for _ in range( self.nPoblation - 1 ):
      self.solution = np.vstack( [ self.solution,  np.random.randint( valCol, size = self.nVertex ) ] )

```
