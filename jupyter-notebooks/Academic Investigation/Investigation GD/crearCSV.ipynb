```
from google.colab import drive
from pathlib import Path
import csv 

drive.mount('/content/gdrive')
```

```
def ls(ruta = Path.cwd()):
    return [arch.name for arch in Path(ruta).iterdir() if arch.is_file()]
```

```
for i in range (1,12):
  ruta = '/content/gdrive/My Drive/DataSets/GdCapitulos/gd'+str(i)+'/'
  lista = ls( ruta )
  out='/content/gdrive/My Drive/DataSets/GdLista/gd'+ str(i) + 'Salida.csv'
  with open(out, 'wt',encoding='utf-8') as file:  
    cap = 0
    headers = ['ncap','npag','nlinea','tocr']
    writer = csv.writer(file, quoting=csv.QUOTE_NONNUMERIC, lineterminator='\n')
    writer.writerow(('ncap','npag','nlinea','tocr','NULL', 'NULL', 'NULL'))
    for arc in lista:
      f = open( ruta + arc, encoding='utf-8')
      nrolinea = 0
      cap += 1
      for linea in f:
        linea = linea.replace(',','\,')
        nrolinea += 1
        if 'Pagina: ' in linea:
          pag = int( str(linea.split()[1]))
          nrolinea -= 1 
        elif len(linea)>2:
          writer.writerow( ( '',cap, pag ,nrolinea, linea.strip() ) )
    f.close()
```
